#### Add a BST (base stat total) column

```python
# long way DONT DO THIS
df["BST"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
```

```python
df.head()
```

```python
# drop column
df = df.drop(columns = "BST")
df.head()
```

```python
# way better way using .iloc
df["BST"] = df.iloc[:, 4:10].sum(axis = 1)
```

```python
df.head()
```

#### Change ordering of columns

```python
# list(df.columns.values) also works
cols = list(df.columns)
cols
```

```python
df = df[cols[0:10] + [cols[-1]] + cols[10:12]]
df.head()
```

### Filtering Data

#### Filter data by their types

```python
df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
```

```python
grass_poison_df = df.loc[(df["Type 1"] == "Grass") & (df["Type 2"] == "Poison") & (df["HP"] > 70)]
grass_poison_df.head()
```

```python
# notice that the index number stayed the same; change it by using reset_index function
grass_poison_df.reset_index()
```

```python
# reset_index saves old index, to remove that, just set drop parameter = True
grass_poison_df.reset_index(drop = True)
```

```python
grass_poison_df.head()
```

```python
# we have not saved this resetting index; we can save this two ways: setting it equal to itself or using the inplace parameter
grass_poison_df.reset_index(drop = True, inplace = True)
grass_poison_df.head()
```

### Filter data by string

```python
# find megas
df.loc[df["Name"].str.contains("Mega")]
```

```python
# find grass or poison types using str.contains()
# possible because we import re at top of page
df.loc[df["Type 1"].str.contains("fire|grass", flags = re.I)]
```

```python
# find pokemon names that begin with "pi"
df.loc[df["Name"].str.contains("^pi[a-z]*", flags = re.I)]
```

```python
# create dummy df just so we can do these changes and not mess up the original
dummy_df = df
dummy_df.head()
```

```python
# lets say i want to change "fire" into just an F
dummy_df.loc[dummy_df["Type 1"] == "Fire", "Type 1"] = "F"
dummy_df.head(10)
```

```python
# lets say i want to make all pokemon with a BST of over 600 legendary
dummy_df.loc[dummy_df["BST"] > 600, "Legendary"] = True
```

```python
# how do we find the mean of all types (so that their type1 or type2 doesnt matter??)
# one way to do it would be:
df[(df["Type 1"] == "Bug") | (df["Type 2"] == "Bug")].mean()
```

```python
# how do we generalize?
unique_types = df["Type 1"].unique()
unique_types
```

```python
# check if we can generalize with a for loop
# commenting out for readability

# for pokemon_type in unique_types:
    # print(pokemon_type)
    # print(df[(df["Type 1"] == pokemon_type) | (df["Type 2"] == pokemon_type)].mean())
    # print("\n")
    
print("Grass")
print(df[(df["Type 1"] == "Grass") | (df["Type 2"] == "Grass")].mean())
print("\n")
```

```python
# check correctedness
df[(df["Type 1"] == "Grass") | (df["Type 2"] == "Grass")].mean()
```

```python
# play with pd.concat
pd.concat([pd.DataFrame(df[(df["Type 1"] == "Grass") | (df["Type 2"] == "Grass")].mean()), 
           pd.DataFrame(df[(df["Type 1"] == "Fire") | (df["Type 2"] == "Fire")].mean())], axis = 1)
```

```python
# initialize empty dataframe for to use pd.concat
df_types = pd.DataFrame()
```

```python
for pokemon_type in unique_types:
    df_types = pd.concat([df_types, df[(df["Type 1"] == pokemon_type) | (df["Type 2"] == pokemon_type)].mean()], axis = 1)
```

```python
df_types
```

```python
types_for_df = np.array([])
for pokemon_type in unique_types:
    types_for_df = np.append(types_for_df, pokemon_type)
```

```python
types_for_df
```

```python
df_types.columns = types_for_df
```
